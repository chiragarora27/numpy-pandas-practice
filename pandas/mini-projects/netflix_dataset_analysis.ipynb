ANALYSIS

```python
def top_movies():
    global df
    filt = df['type'] == 'Movie'
    top_movies = df.loc[filt].nlargest(10, 'points')
    
    movie_titles = top_movies['title']
    movie_points = top_movies['points']

    best_movies_df = pd.concat([movie_titles, movie_points], axis=1)
    print ("The best movies according to points are as follows:\n")
    return best_movies_df   
```

```python
def common_genres():
    global df

    genres_common = df['listed_in'].value_counts().head(5)
    print ("The most common genres are as follows:\n")

    return genres_common
```

```python
def common_countries():
    global df
    print("A large chunk of shows and movies are from the following countries: ")
    return df['country'].value_counts().head(5)
```

```python
def average_per_genre():
    global df
    genre = input("Enter genre: ")
    filt = df['listed_in'].str.contains(genre, case=False)

    print("The average points for the chosen genre are: ")
    return df.loc[filt, 'points'].mean()
```

```python
def common_directors():
    global df
    print ("The most common genres are as follows:\n")
    return df['director'].value_counts().head(5)
```

FACTORED ANALYSIS

```python
def mean_runtime_by_country():
    global df

    country = input('Which country? ')
    
    country_grp = df.loc[df['type'] == 'Movie'].groupby('country')

    print("The average runtime in minutes for the chosen country is:")
    return country_grp['duration'].agg(['mean']).loc[country]
```

```python
def mean_points_by_country():
    global df

    country = input('Which country? ')
    
    country_grp = df.loc[df['type'] == 'Movie'].groupby('country')

    print("The average points in minutes for the chosen country is:")
    return country_grp['points'].agg(['mean']).loc[country]
```

```python
def year_wise_points():
    global df
    year_grp = df.groupby('release_year')
    year_mean = year_grp['points'].mean()
    year_mean.sort_values(ascending=False, inplace=True)
    return year_mean
    
```

```python
def runtime_by_genres_movie():
    global df
    genre_grp = df.loc[df['type'] == 'Movie'].groupby('listed_in')
    genre_points = genre_grp['duration'].mean()
    genre_points.sort_values(ascending=False, inplace=True)
    print("In our data, we can notice that classic movies have the highest runtime,\nmaybe due to their timelessness and cultural impact")
    return genre_points
```

```python
def runtime_by_genres_show():
    global df
    genre_grp = df.loc[df['type'] == 'TV Show'].groupby('listed_in')
    genre_points = genre_grp['duration'].mean()
    genre_points.sort_values(ascending=False, inplace=True)
    print("In our data we can notice that classic and cult TV have the highest runtime,\npossibly due to their higher populatiry and neverending scripts")
    return genre_points
```

RUNTIME AND RATING RELATION ANALYSIS

```python
def runtime_rating_relation_movie():
    global df
    runtime = df.loc[df['type'] == 'Movie', 'duration']
    rating = df.loc[df['type'] == 'Movie', 'points']

    new_df = pd.concat([runtime, rating], axis=1)
    correlation = np.corrcoef(new_df['duration'] , new_df['points'])

    if abs(correlation[0,1]) > 0.70:
        return("The two are somewhat correlated")

    else: 
        return("The two are not related")
```

```python
df['duration_hours'] = df['duration'] / 60 
```

MULTI-FACTOR RANKING SYSTEM 

```python
def ranking_system():
    global df
    # points get 70% weightage in this ranking system.
    # release_year gets 10% weightage
    # category gets 10% weightage
    # country gets 10% weightage

    # these are kind of a bit senseless since i did not have much relevant columns to work with

    coin_points = df['points'] / 10 * 70
    
    df['coin_release_year'] = df['release_year']
    filt1 = (df['release_year'] > 2015)
    filt2 = (df['release_year'] > 2005) & (df['release_year'] <= 2015)
    filt3 = (df['release_year'] > 1995) & (df['release_year'] <= 2005)
    filt4 = (df['release_year'] > 1980) & (df['release_year'] <= 1995)
    filt5 = (df['release_year'] > 1970) & (df['release_year'] <= 1980)
    filt6 = (df['release_year'] <= 1970)

    df.loc[filt1, 'coin_release_year'] = 10
    df.loc[filt2, 'coin_release_year'] = 8
    df.loc[filt3, 'coin_release_year'] = 6
    df.loc[filt4, 'coin_release_year'] = 4
    df.loc[filt5, 'coin_release_year'] = 2
    df.loc[filt6, 'coin_release_year'] = 0


    coin_release_year = df['coin_release_year']
    df = df.drop('coin_release_year', axis=1)

    df['coin_category'] = df['category']
    df['coin_category'] = df['coin_category'].str.replace('medium', '10')
    df['coin_category'] = df['coin_category'].str.replace('long', '5')
    df['coin_category'] = df['coin_category'].str.replace('short', '5')
    
    df['coin_category'] = df['coin_category'].astype(float)

    coin_category = df['coin_category']
    df = df.drop('coin_category', axis=1) 

    df['coin_country'] = 0

    df.loc[df['country'] == 'India', 'coin_country'] = 10
    df.loc[df['country'] == 'United States of America', 'coin_country'] = 8
    df.loc[df['country'] == 'United Kingdom', 'coin_country'] = 6
    df.loc[df['country'] == 'Japan', 'coin_country'] = 4
    df.loc[df['country'] == 'South Korea', 'coin_country'] = 2

    coin_country = df['coin_country']
    df = df.drop('coin_country', axis=1)

    coins = coin_points + coin_category + coin_country

    df['coins'] = coins
    
    df.sort_values(by= 'coins', ascending=False, inplace=True)
    return df
    
```
