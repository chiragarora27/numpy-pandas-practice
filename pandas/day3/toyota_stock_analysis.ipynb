```python
stocks_df['Growth'] = stocks_df['Close'] - stocks_df['Open']
```

```python
month_grp['Growth'].sum().sort_values(ascending=False)
```

```python
print("It is evident that the stock grows most in the month of november and april.")
```

```python
def analyse_year(year):
    max_price = year_grp['High'].max().loc[year]
    min_price = year_grp['Low'].min().loc[year]
    avg_growth = year_grp['Growth'].mean().loc[year]
    avg_close = year_grp['Close'].mean().loc[year]

    return_statement = "The analysis for the year {} is as following:\n1. Highest Stock Price = {}\n2. Lowest Stock Price = {}\n3. Average Growth = {}\n4. Average Close Price = {}".format(year, max_price, min_price, avg_growth, avg_close)

    return return_statement
```

```python
print(analyse_year(2020))
```
