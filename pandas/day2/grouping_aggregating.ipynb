```python
# 66 shows from india on netflix are comedy genre
```

```python
def no_of_true(genre):
    return genre.str.contains('Comedy').sum()
    
    
```

```python
country_grp['genres'].apply(lambda x: x.str.contains('Comedy'))['India'].sum()
```

```python
country_grp['genres'].apply(no_of_true)['India']
```

```python
country_grp['genres'].apply(no_of_true)
```

```python
country_grp['show_id'].count()
```

```python
country_grp['genres'].apply(no_of_true) / country_grp['show_id'].count() 
```

```python
country_grp['genres'].apply(lambda x: x.str.contains('Comedy').sum() / len(x) )
```

```python
country_comedy = country_grp['genres'].apply(no_of_true)
country_comedy
```

```python
country_count = country_grp['show_id'].count()
```

```python
country_count
```

```python
python_df = pd.concat([country_count, country_comedy], axis=1)
```

```python
python_df
```

```python
python_df.rename(columns= {'show_id' : 'total_entries', 'genres' : 'total_comedy'}, inplace=True)
```

```python
python_df['comedy_percent'] = python_df['total_comedy'] / python_df['total_entries'] * 100
```

```python
python_df
```

```python
python_df.sort_values(by= 'comedy_percent', ascending=False, inplace=True)
```
