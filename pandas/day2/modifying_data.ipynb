```python
import pandas as pd
```

```python
df = pd.read_csv('data1.csv')
```

```python
df.rename(columns = {'rating(of 10)' : 'points'}, inplace= True)
```

```python
def reccomendation(rating):
    if rating < 6:
        return 'bad'
    else: 
        return 'good'
```

```python
df['reccomendation'] = df['points'].apply(reccomendation)
```

```python
df['reccomendation']
```

```python
df.shape
```

```python
def time(year):
    if year < 2018:
        return 'old'
    else:
        return 'new'
```

```python
df['gen'] = df['release_year'].apply(time)
```

```python
df.sort_values(by= ['country', 'points'], ascending=[True, False], inplace=True)
```
