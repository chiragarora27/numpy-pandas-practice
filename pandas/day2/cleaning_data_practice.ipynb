```python
import pandas as pd
import numpy as np
```

```python
people = {
    'first': ['Corey', 'Jane', 'John', 'Chris', np.nan, None, 'NA'], 
    'last': ['Schafer', 'Doe', 'Doe', 'Schafer', np.nan, np.nan, 'Missing'], 
    'email': ['CoreyMSchafer@gmail.com', 'JaneDoe@email.com', 'JohnDoe@email.com', None, np.nan, 'Anonymous@email.com', 'NA'],
    'age': ['33', '55', '63', '36', None, None, 'Missing']
}
```

```python
df = pd.DataFrame(people)
```

```python
df
```

```python
df.dropna() # this drops the rows with missing values
```

```python
df.dropna(axis=0, how='any') # these are the default arguements, we can edit them.
```

```python
df.dropna(axis=0, how='all')
```

```python
df.dropna(subset=['email']) #only rows with missing emails are dropped
```

```python
#remember custom missing values like NA and Missing are not caught by dropna, smth else is done to take care of them
```

```python
df.isna()
```

```python
df.replace('NA', np.nan, inplace=True)
df.replace('Missing', np.nan, inplace=True)
```

```python
df.isna()
```

```python
df.fillna(0)
```

```python
df['age'] = df['age'].astype(float)
```

```python
df['age'].mean()
```
